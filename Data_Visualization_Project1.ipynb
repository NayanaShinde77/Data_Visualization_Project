# Explore Data

```python
# 3. Summary Statistics
print(df.describe())     # Count, mean, std, min, quartiles, max
print(df.describe(include='all'))  # Include categorical columns
```

```python
# 5. Unique Values & Counts
for col in df.columns:
    print(f"{col} unique values: {df[col].nunique()}")  # Number of unique values
    print(df[col].unique())                              # Show unique values
```

```python

# 6. Value Counts (Categorical Column)
print(df['Species'].value_counts())
print(df['Species'].value_counts(normalize=True))  # Percentage
```

```python
# 7. Sorting & Selecting
print(df.sort_values(by='SepalLength', ascending=False).head(5))  # Top 5 largest sepal lengths
print(df.nlargest(5, 'SepalWidth'))  # Top 5 largest petal widths
print(df.nsmallest(5, 'PetalLength'))  # Top 5 smallest petal lengths
```

```python
# 8. Correlations
numeric_df = df.select_dtypes(include=np.number)  # Select only numeric columns
print(numeric_df.corr())                         # Pearson correlation
print(numeric_df.corr(method='spearman'))       # Spearman correlation
print(numeric_df.corr(method='kendall'))        # Kendall correlation
```

```python
# 9. Grouping & Aggregation
print(df.groupby('Species').mean())      # Mean of numeric columns by species
print(df.groupby('Species').median())    # Median by species
print(df.groupby('Species').agg(['mean','median','std']))  # Multiple aggregations
```

```python
# 1. Boxplot
sns.boxplot(x='Species', y='SepalLength', data=df)
plt.title("Boxplot of Sepal Length by Species")
plt.show()
```

```python
# 2. Countplot
sns.countplot(x='Species', data=df)
plt.title("Count of Each Species")
plt.show()
```

```python
# 3. Lineplot (mean sepal_length per species)
df.groupby('Species')['SepalLength'].mean().plot(kind='line', marker='o')
plt.title("Mean Sepal Length per Species")
plt.ylabel("Sepal Length")
plt.show()
```

```python
# 4. Scatter plot
sns.scatterplot(x='SepalLength', y='SepalWidth', hue='Species', data=df)
plt.title("Sepal Length vs Width")
plt.show()
```

```python
#5. Heatmap(Correlation)
sns.heatmap(df.drop(columns=['Species']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 6. Pie Chart
df['Species'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Species Distribution")
plt.ylabel('')
plt.show()
```

```python
# 7. Bar Plot
df.groupby('Species')['PetalLength'].mean().plot(kind='bar', color='skyblue')
plt.title("Mean Petal Length per Species")
plt.ylabel("Petal Length")
plt.show()
```

```python
# 8. Violin Plot
sns.violinplot(x='Species', y='PetalWidth', data=df)
plt.title("Violin Plot of Petal Width by Species")
plt.show()
```

```python
# 9. KDE Plot
for spec in df['Species'].unique():
    sns.kdeplot(df[df['Species']==spec]['SepalLength'], label=spec)
plt.title("KDE of Sepal Length by Species")
plt.show()
```

```python
# 10. Strip Plot
sns.stripplot(x='Species', y='PetalLength', data=df, jitter=True)
plt.title("Strip Plot of Petal Length by Species")
plt.show()
```

```python
# 11. Pairplot
import warnings
warnings.filterwarnings("ignore")

sns.pairplot(df, hue='Species', diag_kind='kde')
plt.show()
```

```python
# Jointplot
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df, kind='scatter', hue='Species')
plt.show()
```

```python
# 13. Hexbin (scatter density)
plt.hexbin(df['SepalLength'], df['SepalWidth'], gridsize=20, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel("SepalLenthCm")
plt.ylabel("SepalWidthCm")
plt.title("Hexbin Plot of Sepal Dimensions")
plt.show()
```

```python
# 14. Boxen Plot
sns.boxenplot(x='Species', y='PetalLength', data=df)
plt.title("Boxen Plot of Petal Length by Species")
plt.show()
```

```python
# 15. Catplot
sns.catplot(x='Species', y='SepalWidth', kind='box', data=df)
plt.show()
```

```python
# 16. Lmplot
sns.lmplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=df)
plt.title("Linear Relationship Sepal vs Petal Length")
plt.show()
```
